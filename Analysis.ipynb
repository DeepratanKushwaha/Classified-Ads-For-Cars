## create a new dataframe called clean_used_cars from used_cars with the following conditions:
    1. Drop the columns with more than 50% missing values 
    2. The manufacture year between 2000 and 2017 including 2000 and 2017 
    3. Both maker and model exist in the row 
    4. The price range is from 3000 to 2000,000 (3000 ≤ price ≤ 2000,000)
    5. Remove any price you singled out in Step 3 (ie a price that repeats too frequently for a random set of ads).

```python
used_cars
```

## 1. Drop the columns with more than 50% missing values

color_slug & fuel_type are the columns which have more than 50% missing values

```python
clean_used_cars = used_cars.drop(['color_slug', 'fuel_type'], axis=1).copy()
clean_used_cars
```

##    2. The manufacture year between 2000 and 2017 including 2000 and 2017  

```python
clean_used_cars = clean_used_cars[(clean_used_cars.manufacture_year >= 2000) & (clean_used_cars.manufacture_year <= 2017)]
clean_used_cars
```

## 3. Both maker and model exist in the row 

```python
clean_used_cars.maker.isna().sum()
```

```python
clean_used_cars.model.isna().sum()
```

```python
clean_used_cars = clean_used_cars.dropna(axis=0, subset=['maker', 'model'])
clean_used_cars
```

```python
clean_used_cars.maker.isna().sum()
```

```python
clean_used_cars.model.isna().sum()
```

## 4. The price range is from 3000 to 2000,000 (3000 ≤ price ≤ 2000,000)

```python
clean_used_cars = clean_used_cars[(clean_used_cars.price_eur >= 3000) & (clean_used_cars.price_eur <= 2000000)]
clean_used_cars
```

## 5. Remove any price you singled out in Step 3 (ie a price that repeats too frequently for a random set of ads)

## find the make and model for the cars with the top 10 highest average price

```python
clean_used_cars.groupby(['maker', 'model'])['price_eur'].value_counts()
```

```python
clean_used_cars.groupby(['maker', 'model']).agg({'price_eur':'mean'}).reset_index().sort_values('price_eur', ascending=False).head(10)
```

## find the make and model for the cars with the top 10 lowest average price

```python
clean_used_cars.groupby(['maker', 'model']).agg({'price_eur':'mean'}).reset_index().sort_values('price_eur', ascending=False).tail(10)
```

## recommend top five maker and model for Economic segment customers (Top five manufacturers in the 3000 to 20,000 price range;3000≤price<20,000) - based on the top average price

```python
data = clean_used_cars.groupby(['maker', 'model']).agg({'price_eur':'mean'}).reset_index()
economic_segment = data[(data['price_eur'] >= 3000) & (data['price_eur'] <= 20000)]
economic_segment = economic_segment.sort_values('price_eur', ascending=False, ignore_index=True)[:5]
economic_segment
```

## recommend top five make and model for Intermediate segment customers (Top five manufacturers in the 20,000 to 300,000 price range; 3000≤price<20,000) - based on the top average price

```python
intermediate_segment = data[(data['price_eur'] >= 20000) & (data['price_eur'] <= 300000)]
intermediate_segment = intermediate_segment.sort_values('price_eur', ascending=False, ignore_index=True)[:5]
intermediate_segment
```

## recommend the top five make and model for the Luxury segment customers (Top five manufacturers in the 300,000 to 2000,000 price range; 300,000≤price<2000,000) - based on the top average price

```python
luxury_segment = data[(data['price_eur'] >= 300000) & (data['price_eur'] <= 2000000)]
luxury_segment = luxury_segment.sort_values('price_eur', ascending=False, ignore_index=True)[:5]
luxury_segment
```
